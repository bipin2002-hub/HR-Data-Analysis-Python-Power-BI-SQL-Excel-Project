# KPI's

```python
 #Total Employees 
Total_Emp=df["Employee Number"].value_counts().sum()
print("Total Employees :", Total_Emp)
```

```python
# Attrition Count
Attrition_Count=df[df["Attrition"] == "Yes"].value_counts().sum()
print("Attrition :",Attrition_Count)
```

```python
# Average Age of Employee
Avg_age=df["Age"].mean()
print("Average Age of Employee:",Avg_age)
```

```python
# Active Employees 
Active_emp=df[df["CF_attrition label"]=="Current Employees"].value_counts().sum()
print("Active Employees",Active_emp)
```

```python
#Attrition Rate %
Attrition_Rate=Attrition_Count*100/Total_Emp
print("Attrition Rate:",Attrition_Rate,"%")
```
